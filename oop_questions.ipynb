```python
'''Parking Lot'''
class Vehicle:
    def __init__(self, number, vehicle_type):
        self.number = number
        self.vehicle_type = vehicle_type

class ParkingLot:
    rates = {
        "car": 20,
        "bike": 10,
        "truck": 30
    }
    def __init__(self, total_spots):
        self.total_spots = total_spots
        self.available_spots = total_spots
        self.parked_vehicles = {}

    def park_vehicle(self):
        if self.available_spots == 0:
            print("❌ Parking Full!")
            return

        number = input("Enter vehicle number: ").upper()
        vehicle_type = input("Enter vehicle type (Car/Bike/Truck): ").lower()

        if vehicle_type not in ParkingLot.rates:
            print("❌ Invalid vehicle type!")
            return

        if number in self.parked_vehicles:
            print("❌ Vehicle already parked!")
            return

        vehicle = Vehicle(number, vehicle_type)
        self.parked_vehicles[number] = vehicle
        self.available_spots -= 1

        print("✅ Vehicle parked successfully.")

    def exit_vehicle(self):
        number = input("Enter vehicle number to exit: ").upper()

        if number not in self.parked_vehicles:
            print("❌ Vehicle not found!")
            return

        hours = int(input("Enter number of hours parked: "))

        if hours <= 0:
            print("❌ Invalid hours!")
            return

        vehicle = self.parked_vehicles[number]
        rate = ParkingLot.rates[vehicle.vehicle_type]
        fee = hours * rate

        del self.parked_vehicles[number]
        self.available_spots += 1

        print("✅ Vehicle exited successfully.")
        print("Parking Duration:", hours, "hour(s)")
        print("Total Fee: ₹", fee)

    def show_status(self):
        print("\n----- Parking Status -----")
        print("Total Spots:", self.total_spots)
        print("Available Spots:", self.available_spots)
        print("Parked Vehicles:", len(self.parked_vehicles))
        print("--------------------------\n")

total = int(input("Enter total parking spots: "))
lot = ParkingLot(total)

while True:
    print("\n1. Park Vehicle")
    print("2. Exit Vehicle")
    print("3. Show Status")
    print("4. Exit System")

    choice = input("Enter your choice: ")

    if choice == "1":
        lot.park_vehicle()
    elif choice == "2":
        lot.exit_vehicle()
    elif choice == "3":
        lot.show_status()
    elif choice == "4":
        print("🚗 Thank you for using Parking System!")
        break
    else:
        print("❌ Invalid choice. Try again.")
```

```python
'''BILLING SYSTEM FOR RETAIL STORE'''
class Product:
    def __init__(self, pid, name, price):
        self.pid = pid
        self.name = name
        self.price = price


class BillingSystem:
    def __init__(self):
        self.products = {}
        self.cart = []
        self.transactions = []

    def add_product(self):
        pid = input("Enter Product ID: ")
        name = input("Enter Product Name: ")
        price = float(input("Enter Price: "))
        self.products[pid] = Product(pid, name, price)
        print("Product added successfully.")

    def scan_product(self):
        pid = input("Enter Product ID to scan: ")
        if pid in self.products:
            qty = int(input("Enter quantity: "))
            self.cart.append((self.products[pid], qty))
            print("Product added to cart.")
        else:
            print("Product not found.")

    def generate_bill(self):
        if not self.cart:
            print("Cart is empty.")
            return

        total = 0
        print("\n------ BILL ------")
        print("Product\tQty\tPrice\tTotal")

        for product, qty in self.cart:
            amount = product.price * qty
            total += amount
            print(product.name, "\t", qty, "\t", product.price, "\t", amount)

        discount = float(input("Enter discount %: "))
        discount_amount = total * discount / 100
        final_total = total - discount_amount

        print("\nSubtotal:", total)
        print("Discount:", discount_amount)
        print("Final Amount:", final_total)

        self.transactions.append(final_total)
        self.cart.clear()

    def show_transactions(self):
        print("\nTransaction History")
        for i, t in enumerate(self.transactions, 1):
            print("Transaction", i, ":", t)


system = BillingSystem()

while True:
    print("\n1.Add Product")
    print("2.Scan Product")
    print("3.Generate Bill")
    print("4.View Transactions")
    print("5.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        system.add_product()

    elif choice == "2":
        system.scan_product()

    elif choice == "3":
        system.generate_bill()

    elif choice == "4":
        system.show_transactions()

    elif choice == "5":
        break
```

```python
class Movie:
    def __init__(self, name, showtime, seats):
        self.name = name
        self.showtime = showtime
        self.seats = seats
        self.booked_seats = []

    def show_details(self):
        print("\nMovie:", self.name)
        print("Showtime:", self.showtime)
        print("Available Seats:", self.seats - len(self.booked_seats))

    def book_seat(self):
        seat = int(input("Enter seat number: "))
        if seat in self.booked_seats:
            print("Seat already booked.")
        elif seat > self.seats or seat <= 0:
            print("Invalid seat number.")
        else:
            self.booked_seats.append(seat)
            print("Seat booked successfully.")
            print("Booking Confirmation")
            print("Movie:", self.name)
            print("Seat Number:", seat)
            print("Showtime:", self.showtime)


class BookingSystem:
    def __init__(self):
        self.movies = []

    def add_movie(self):
        name = input("Enter movie name: ")
        showtime = input("Enter showtime: ")
        seats = int(input("Enter total seats: "))
        self.movies.append(Movie(name, showtime, seats))

    def show_movies(self):
        for i, m in enumerate(self.movies):
            print(i + 1, m.name, "-", m.showtime)

    def book_ticket(self):
        self.show_movies()
        choice = int(input("Select movie number: ")) - 1
        if 0 <= choice < len(self.movies):
            self.movies[choice].show_details()
            self.movies[choice].book_seat()
        else:
            print("Invalid choice")


system = BookingSystem()

while True:
    print("\n1.Add Movie")
    print("2.Show Movies")
    print("3.Book Ticket")
    print("4.Exit")

    ch = input("Enter choice: ")

    if ch == "1":
        system.add_movie()
    elif ch == "2":
        system.show_movies()
    elif ch == "3":
        system.book_ticket()
    elif ch == "4":
        break
```

```python
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        self.marks = []

    def enter_marks(self):
        n = int(input("Enter number of subjects: "))
        for i in range(n):
            m = float(input(f"Enter marks for subject {i+1}: "))
            self.marks.append(m)

    def calculate_gpa(self):
        avg = sum(self.marks) / len(self.marks)

        if avg >= 90:
            return 10
        elif avg >= 80:
            return 9
        elif avg >= 70:
            return 8
        elif avg >= 60:
            return 7
        elif avg >= 50:
            return 6
        else:
            return 5

    def display_result(self):
        print("\nStudent Name:", self.name)
        print("Roll No:", self.roll)
        print("Marks:", self.marks)
        print("GPA:", self.calculate_gpa())


class ResultSystem:
    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter student name: ")
        roll = input("Enter roll number: ")
        s = Student(name, roll)
        s.enter_marks()
        self.students.append(s)

    def show_results(self):
        for s in self.students:
            s.display_result()


system = ResultSystem()

while True:
    print("\n1.Add Student Marks")
    print("2.Show Results")
    print("3.Exit")

    ch = input("Enter choice: ")

    if ch == "1":
        system.add_student()
    elif ch == "2":
        system.show_results()
    elif ch == "3":
        break
```

```python
class Task:
    def __init__(self, title, deadline, priority, category):
        self.title = title
        self.deadline = deadline
        self.priority = priority
        self.category = category

    def display(self):
        print(f"\nTask: {self.title}")
        print(f"Deadline: {self.deadline}")
        print(f"Priority: {self.priority}")
        print(f"Category: {self.category}")
        print("----------------------")

class TodoManager:
    def __init__(self):
        self.tasks = []

    def add_task(self):
        title = input("Enter task title: ")
        deadline = input("Enter deadline (YYYY-MM-DD): ")
        priority = input("Enter priority (High/Medium/Low): ")
        category = input("Enter category/project: ")

        task = Task(title, deadline, priority, category)
        self.tasks.append(task)

        print("✅ Task added successfully!")

    def show_tasks(self):
        if not self.tasks:
            print("No tasks available.")
            return

        print("\nAll Tasks:")
        for task in self.tasks:
            task.display()

    def filter_category(self):
        cat = input("Enter category to filter: ")
        found = False

        for task in self.tasks:
            if task.category.lower() == cat.lower():
                task.display()
                found = True

        if not found:
            print("No tasks found in this category.")

    def filter_priority(self):
        pr = input("Enter priority to filter (High/Medium/Low): ")
        found = False

        for task in self.tasks:
            if task.priority.lower() == pr.lower():
                task.display()
                found = True

        if not found:
            print("No tasks found with this priority.")


todo = TodoManager()

while True:
    print("\n===== TO-DO LIST MANAGER =====")
    print("1. Add Task")
    print("2. Show All Tasks")
    print("3. Filter by Category")
    print("4. Filter by Priority")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        todo.add_task()

    elif choice == "2":
        todo.show_tasks()

    elif choice == "3":
        todo.filter_category()

    elif choice == "4":
        todo.filter_priority()

    elif choice == "5":
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
'''Password Manager'''
import base64

class PasswordEntry:
    def __init__(self, website, username, password):
        self.website = website
        self.username = username
        self.password = base64.b64encode(password.encode()).decode()

    def get_password(self):
        return base64.b64decode(self.password.encode()).decode()

class PasswordManager:
    def __init__(self):
        self.passwords = {}

    def add_password(self):
        website = input("Enter website: ")
        username = input("Enter username: ")
        password = input("Enter password: ")
        self.passwords[website] = PasswordEntry(website, username, password)
        print("Password stored successfully.")

    def edit_password(self):
        website = input("Enter website to update: ")
        if website in self.passwords:
            new_pass = input("Enter new password: ")
            self.passwords[website].password = base64.b64encode(new_pass.encode()).decode()
            print("Password updated.")
        else:
            print("Website not found.")

    def retrieve_password(self):
        website = input("Enter website to retrieve: ")
        if website in self.passwords:
            entry = self.passwords[website]
            print("Username:", entry.username)
            print("Password:", entry.get_password())
        else:
            print("No record found.")


pm = PasswordManager()

while True:
    print("\n1.Add Password 2.Edit Password 3.Retrieve Password 4.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        pm.add_password()
    elif choice == "2":
        pm.edit_password()
    elif choice == "3":
        pm.retrieve_password()
    elif choice == "4":
        break
```

```python
'''Freelancer Marketplace'''
class Freelancer:
    def __init__(self, name, skill):
        self.name = name
        self.skill = skill


class Client:
    def __init__(self, name):
        self.name = name


class Project:
    def __init__(self, title, budget):
        self.title = title
        self.budget = budget
        self.freelancer = None


freelancers = []
clients = []
projects = []

while True:
    print("\n1.Register Freelancer")
    print("2.Register Client")
    print("3.Create Project")
    print("4.Assign Freelancer")
    print("5.Process Payment")
    print("6.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Freelancer name: ")
        skill = input("Skill: ")
        freelancers.append(Freelancer(name, skill))
        print("Freelancer registered.")

    elif choice == "2":
        name = input("Client name: ")
        clients.append(Client(name))
        print("Client registered.")

    elif choice == "3":
        title = input("Project title: ")
        budget = float(input("Budget: "))
        projects.append(Project(title, budget))
        print("Project created.")

    elif choice == "4":
        pname = input("Enter project title: ")
        fname = input("Enter freelancer name: ")
        for p in projects:
            if p.title == pname:
                for f in freelancers:
                    if f.name == fname:
                        p.freelancer = f
                        print("Freelancer assigned.")
                        break

    elif choice == "5":
        pname = input("Enter project title: ")
        for p in projects:
            if p.title == pname and p.freelancer:
                print("Payment of", p.budget, "paid to", p.freelancer.name)

    elif choice == "6":
        break
```

```python
'''warehouse automation'''
class Product:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity


class Warehouse:
    def __init__(self):
        self.inventory = {}

    def add_goods(self):
        name = input("Product name: ")
        qty = int(input("Quantity: "))
        if name in self.inventory:
            self.inventory[name].quantity += qty
        else:
            self.inventory[name] = Product(name, qty)
        print("Goods added.")

    def remove_goods(self):
        name = input("Product name: ")
        qty = int(input("Quantity: "))
        if name in self.inventory and self.inventory[name].quantity >= qty:
            self.inventory[name].quantity -= qty
            print("Goods removed.")
        else:
            print("Insufficient stock.")

    def report(self):
        print("\nInventory Report")
        for p in self.inventory.values():
            print(p.name, ":", p.quantity)

    def forecast(self):
        name = input("Enter product name: ")
        if name in self.inventory:
            print("Forecast demand:", self.inventory[name].quantity * 2)


wh = Warehouse()

while True:
    print("\n1.Add Goods 2.Remove Goods 3.Inventory Report 4.Forecast Demand 5.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        wh.add_goods()
    elif choice == "2":
        wh.remove_goods()
    elif choice == "3":
        wh.report()
    elif choice == "4":
        wh.forecast()
    elif choice == "5":
        break
```

```python
''' CRM(Customer Relationship Manager)'''
class Customer:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.logs = []


class CRM:
    def __init__(self):
        self.customers = []

    def add_customer(self):
        name = input("Customer name: ")
        email = input("Customer email: ")
        self.customers.append(Customer(name, email))
        print("Customer added.")

    def add_log(self):
        name = input("Customer name: ")
        message = input("Communication message: ")
        for c in self.customers:
            if c.name == name:
                c.logs.append(message)
                print("Log added.")

    def show_customers(self):
        for c in self.customers:
            print("\nName:", c.name)
            print("Email:", c.email)
            print("Logs:", c.logs)


crm = CRM()

while True:
    print("\n1.Add Customer 2.Add Communication Log 3.Show Customers 4.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        crm.add_customer()
    elif choice == "2":
        crm.add_log()
    elif choice == "3":
        crm.show_customers()
    elif choice == "4":
        break
```
