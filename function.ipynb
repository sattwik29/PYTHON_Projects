```python
def reverse_str(str1):
  return str1[::-1]
print(reverse_str("ANKIT"))

```

```python
def is_palindrome(str):
  return str == str[::-1]
print(is_palindrome("ATTA"))
```

```python
def factorial(n):
  if n==0 or n==1:
    return 1
  return n*factorial(n-1)
print(factorial(6))
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count
print(count_vowels("ANKIT sahu"))
```

```python
def find_largest(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest
print(find_largest([12,3,4,68,34,69,]))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(17))
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
print(sum_of_digits(4352))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
fibonacci(9)
```

```python
def remove_duplicates(lst):
    new_list = []
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    return new_list
print(remove_duplicates([1,2,2,3,4,3,4,8,9,0,5,8,6,2,7,7,5]))
```

```python
def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count
print(string_length("ankit sahu"))
```
