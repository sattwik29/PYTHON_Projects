```python
accounts = {
    "1001": {"pin": "1234", "balance": 5000},
    "1002": {"pin": "5678", "balance": 3000},
    "1003": {"pin": "0000", "balance": 7000}
}

def deposit(account_no, amount):
    if amount > 0:
        accounts[account_no]["balance"] += amount
        print(f"Deposit successful. New balance: ₹{accounts[account_no]['balance']}")
    else:
        print("Deposit amount must be positive.")

def withdraw(account_no, amount):
    if amount > 0:
        if accounts[account_no]["balance"] >= amount:
            accounts[account_no]["balance"] -= amount
            print(f"Withdrawal successful. New balance: ₹{accounts[account_no]['balance']}")
        else:
            print("Insufficient funds.")
    else:
        print("Withdrawal amount must be positive.")

def check_balance(account_no):
    print(f"Current Balance: ₹{accounts[account_no]['balance']}")

print("===== Welcome to Multi-Account ATM =====")

# Login Section
account_no = input("Enter your Account Number: ")

if account_no in accounts:
    attempts = 0
    max_attempts = 3

    while attempts < max_attempts:
        pin = input("Enter your PIN: ")
        if pin == accounts[account_no]["pin"]:
            print("Login successful!")
            break
        else:
            attempts += 1
            print(f"Incorrect PIN. {max_attempts - attempts} attempts left.")
    else:
        print("Too many incorrect attempts. Account blocked.")
        exit()

    # ATM Menu
    while True:
        print("\n1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Logout")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            check_balance(account_no)

        elif choice == "2":
            try:
                amount = float(input("Enter amount to deposit: "))
                deposit(account_no, amount)
            except ValueError:
                print("Invalid amount.")

        elif choice == "3":
            try:
                amount = float(input("Enter amount to withdraw: "))
                withdraw(account_no, amount)
            except ValueError:
                print("Invalid amount.")

        elif choice == "4":
            print("Logged out successfully. Thank you!")
            break

        else:
            print("Invalid choice. Please select 1-4.")

else:
    print("Account number not found.")
```
