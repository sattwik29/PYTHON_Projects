```python
rooms = {
    101: 'none',
    102: 'none',
    103: 'none',
    104: 'none'
}

occupied = []

while True:
    print("\n1. Allocate room")
    print("2. Vacant rooms")
    print("3. Room info")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        name = input("Enter the name: ")
        room = int(input("Enter the room number: "))

        if room in rooms:
            if room in occupied:
                print("Room already occupied.")
            else:
                rooms[room] = name
                occupied.append(room)
                print("Room allocated successfully.")
        else:
            print("Invalid room number.")

    elif choice == 2:
        print("Vacant rooms:")
        vacant = False
        for room, status in rooms.items():
            if status == 'none':
                print(room)
                vacant = True
        if not vacant:
            print("No vacant rooms available.")

    elif choice == 3:
        room = int(input("Enter the room number: "))
        if room in rooms:
            if rooms[room] == 'none':
                print(f"Room {room} is vacant.")
            else:
                print(f"Room {room} is occupied by {rooms[room]}.")
        else:
            print("Invalid room number.")

    elif choice == 4:
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Try again.")
```
