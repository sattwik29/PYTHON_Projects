```python
names = []
rollnos = []
subjects = []
subject_marks = []
lab_marks = []
assignment_marks = []
attendance_list = []
totals = []
grades = []

def calculate_grade(total):
    if total > 90:
        return "O"
    elif total > 80:
        return "E"
    elif total > 70:
        return "A"
    elif total > 60:
        return "B"
    elif total > 50:
        return "C"
    elif total >= 40:
        return "D"
    else:
        return "FAIL"

def add_student():
    name = input("Enter Student Name: ")
    roll = input("Enter Roll Number: ")
    subject = input("Enter Subject Name: ")

    print("\nEnter marks with following limits:")
    print("Subject Exam Marks (Max 60)")
    print("Lab Project Marks (Max 20)")
    print("Assignment Marks (Max 10)")
    print("Attendance Marks (Max 10)")

    sub_mark = float(input("Enter Subject Exam Marks: "))
    if sub_mark > 60 or sub_mark < 0:
        print("❌ Subject Marks cannot exceed 60 or be negative!")
        return

    lab_mark = float(input("Enter Lab Project Marks: "))
    if lab_mark > 20 or lab_mark < 0:
        print("❌ Lab Marks cannot exceed 20 or be negative!")
        return

    assign_mark = float(input("Enter Assignment Marks: "))
    if assign_mark > 10 or assign_mark < 0:
        print("❌ Assignment Marks cannot exceed 10 or be negative!")
        return

    attendance = float(input("Enter Attendance Marks: "))
    if attendance > 10 or attendance < 0:
        print("❌ Attendance Marks cannot exceed 10 or be negative!")
        return

    total = sub_mark + lab_mark + assign_mark + attendance
    grade = calculate_grade(total)

    names.append(name)
    rollnos.append(roll)
    subjects.append(subject)
    subject_marks.append(sub_mark)
    lab_marks.append(lab_mark)
    assignment_marks.append(assign_mark)
    attendance_list.append(attendance)
    totals.append(total)
    grades.append(grade)

    print("✅ Student Record Added Successfully!\n")

def display_students():
    if len(names) == 0:
        print("No records found!")
        return

    print("\n" + "="*90)
    print("Name\tRoll\tSubject\tTotal\tGrade")
    print("="*90)

    for i in range(len(names)):
        print(f"{names[i]}\t{rollnos[i]}\t{subjects[i]}\t{totals[i]}\t{grades[i]}")

    print("="*90)

def search_student():
    roll = input("Enter Roll Number to Search: ")

    for i in range(len(rollnos)):
        if rollnos[i] == roll:
            print("\nStudent Found!")
            print("Name:", names[i])
            print("Subject:", subjects[i])
            print("Subject Marks:", subject_marks[i])
            print("Lab Marks:", lab_marks[i])
            print("Assignment Marks:", assignment_marks[i])
            print("Attendance:", attendance_list[i])
            print("Total:", totals[i])
            print("Grade:", grades[i])
            return

    print("❌ Student Not Found!")

while True:
    print("\n🎓 STUDENT GRADEBOOK MENU")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Search Student by Roll Number")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        display_students()

    elif choice == "3":
        search_student()

    elif choice == "4":
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice!")
```
