```python
students = {}
courses = {}
enrollments = {}

while True:
    print("\n===== STUDENT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. Add Course")
    print("3. Enroll Student")
    print("4. Drop Course")
    print("5. Assign Grade")
    print("6. Mark Attendance")
    print("7. View Student Report")
    print("8. Course Wise Report")
    print("9. Defaulters List")
    print("10. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        sec = input("Enter Section: ")

        students[roll] = {"name": name, "sec": sec}
        enrollments[roll] = {}
        print("Student added successfully!")

    elif choice == 2:
        course_id = input("Enter Course ID: ")
        limit = int(input("Enter Course Limit: "))
        attendance_cond = int(input("Enter Attendance Condition (%): "))

        courses[course_id] = {
            "limit": limit,
            "attendance_cond": attendance_cond,
            "enrolled": 0
        }
        print("Course added successfully!")

    elif choice == 3:
        roll = input("Enter Roll No: ")
        course_id = input("Enter Course ID: ")

        if roll in students and course_id in courses:
            if courses[course_id]["enrolled"] < courses[course_id]["limit"]:
                enrollments[roll][course_id] = {
                    "attendance": 0,
                    "grade": None
                }
                courses[course_id]["enrolled"] += 1
                print("Enrollment successful!")
            else:
                print("Course limit reached!")
        else:
            print("Invalid Roll No or Course ID!")

    elif choice == 4:
        roll = input("Enter Roll No: ")
        course_id = input("Enter Course ID: ")

        if roll in enrollments and course_id in enrollments[roll]:
            del enrollments[roll][course_id]
            courses[course_id]["enrolled"] -= 1
            print("Course dropped successfully!")
        else:
            print("Enrollment not found!")

    elif choice == 5:
        roll = input("Enter Roll No: ")
        course_id = input("Enter Course ID: ")
        grade = input("Enter Grade: ")

        if roll in enrollments and course_id in enrollments[roll]:
            enrollments[roll][course_id]["grade"] = grade
            print("Grade assigned successfully!")
        else:
            print("Enrollment not found!")

    elif choice == 6:
        roll = input("Enter Roll No: ")
        course_id = input("Enter Course ID: ")
        attendance = int(input("Enter Attendance %: "))

        if roll in enrollments and course_id in enrollments[roll]:
            enrollments[roll][course_id]["attendance"] = attendance
            print("Attendance updated!")
        else:
            print("Enrollment not found!")

    elif choice == 7:
        roll = input("Enter Roll No: ")

        if roll in students:
            print("\n--- Student Report ---")
            print("Name:", students[roll]["name"])
            print("Section:", students[roll]["sec"])

            for course_id, data in enrollments[roll].items():
                print(f"\nCourse: {course_id}")
                print("Attendance:", data["attendance"], "%")
                print("Grade:", data["grade"])
        else:
            print("Student not found!")

    elif choice == 8:
        course_id = input("Enter Course ID: ")

        if course_id in courses:
            print("\n--- Course Wise Report ---")
            for roll, course_data in enrollments.items():
                if course_id in course_data:
                    print(f"Roll: {roll}, Name: {students[roll]['name']}, "
                          f"Attendance: {course_data[course_id]['attendance']}%, "
                          f"Grade: {course_data[course_id]['grade']}")
        else:
            print("Course not found!")

    elif choice == 9:
        print("\n--- Defaulters List ---")
        for roll, course_data in enrollments.items():
            for course_id, data in course_data.items():
                required = courses[course_id]["attendance_cond"]
                if data["attendance"] < required:
                    print(f"Roll: {roll}, Name: {students[roll]['name']}, "
                          f"Course: {course_id}, Attendance: {data['attendance']}%")

    elif choice == 10:
        print("Exiting program...")
        break

    else:
        print("Invalid choice!")
```

```python
products = [
    {"id": 1, "name": "Laptop", "type": "Electronics", "mrp": 65000},
    {"id": 2, "name": "Smartphone", "type": "Electronics", "mrp": 25500},
    {"id": 3, "name": "Shirt", "type": "Clothing", "mrp": 1500},
    {"id": 4, "name": "Jeans", "type": "Clothing", "mrp": 2500},
    {"id": 5, "name": "Rice Bag", "type": "Grocery", "mrp": 1500}
]
cart = []

def calculate_discount(product):
    if product["type"] == "Electronics":
        return product["mrp"] * 0.10
    elif product["type"] == "Clothing":
        return product["mrp"] * 0.20
    elif product["type"] == "Grocery":
        return product["mrp"] * 0.05
    else:
        return 0
def show_products():
    print("\nAvailable Products:")
    for product in products:
        print(f'{product["id"]}. {product["name"]} ({product["type"]}) - ₹{product["mrp"]}')

def add_to_cart():
    show_products()
    pid = int(input("Enter product ID: "))
    found = False
    for product in products:
        if product["id"] == pid:
            cart.append(product)
            print(f'{product["name"]} added to the cart.')
            found = True
            break
    if not found:
        print("Invalid product ID.")
def remove_from_cart():
    if not cart:
        print("Cart is empty!")
        return

    print("\nItems in Cart:")
    for i, item in enumerate(cart):
        print(f"{i+1}. {item['name']} - ₹{item['mrp']}")

    index = int(input("Enter item number to remove: ")) - 1
    if 0 <= index < len(cart):
        removed = cart.pop(index)
        print(removed["name"], "removed from cart.")
    else:
        print("Invalid choice!")


def checkout():
    if not cart:
        print("Cart is empty!")
        return
    print("\nEnter Customer Details")
    customer_name = input("Enter Name: ")
    mobile_number = input("Enter Mobile Number: ")

    print("\n========== BILL ==========")
    print("Customer Name :", customer_name)
    print("Mobile Number :", mobile_number)
    print("---------------------------")

    total = 0
    total_discount = 0
    for item in cart:
        discount = calculate_discount(item)
        final_price = item["mrp"] - discount

        print(f'{item["name"]} - ₹{item["mrp"]} | Discount: ₹{discount:.2f} | Final: ₹{final_price:.2f}')

        total += final_price
        total_discount += discount

    print("---------------------------")
    print(f"Total Discount: ₹{total_discount:.2f}")
    print(f"Total Amount to Pay: ₹{total:.2f}")
    print("===========================")
    print("Thank You for Shopping! ")

while True:
    print("\n1. Show Products")
    print("2. Add to Cart")
    print("3. Remove from Cart")
    print("4. Checkout")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        show_products()
    elif choice == 2:
        add_to_cart()
    elif choice == 3:
        remove_from_cart()
    elif choice == 4:
        checkout()
    elif choice == 5:
        print("Exiting...")
        break
    else:
        print("Invalid choice!")
```

```python
products = [
    {"id": 1, "name": "Laptop", "type": "Electronics", "mrp": 65000, "stock": 5},
    {"id": 2, "name": "Smartphone", "type": "Electronics", "mrp": 25500, "stock": 10},
    {"id": 3, "name": "Shirt", "type": "Clothing", "mrp": 1500, "stock": 20},
    {"id": 4, "name": "Jeans", "type": "Clothing", "mrp": 2500, "stock": 15},
    {"id": 5, "name": "Rice Bag", "type": "Grocery", "mrp": 1500, "stock": 25}
]

cart = []
VALID_COUPON = "AK10"

def calculate_discount(product):
    if product["type"] == "Electronics":
        return product["mrp"] * 0.10
    elif product["type"] == "Clothing":
        return product["mrp"] * 0.20
    elif product["type"] == "Grocery":
        return product["mrp"] * 0.05
    return 0

def show_products():
    print("\nAvailable Products:")
    print("-" * 50)
    for product in products:
        print(f'{product["id"]}. {product["name"]} | {product["type"]} | ₹{product["mrp"]} | Stock: {product["stock"]}')
    print("-" * 50)

def search_product():
    keyword = input("Enter product name to search: ").lower()
    found = False
    for product in products:
        if keyword in product["name"].lower():
            print(f'Found: {product["name"]} - ₹{product["mrp"]}')
            found = True
    if not found:
        print("No product found!")

def add_to_cart():
    show_products()
    try:
        pid = int(input("Enter product ID: "))
        quantity = int(input("Enter quantity: "))
    except ValueError:
        print("Invalid input!")
        return

    for product in products:
        if product["id"] == pid:
            if quantity <= 0:
                print("Invalid quantity!")
                return
            if quantity > product["stock"]:
                print("Not enough stock available!")
                return

            cart.append({
                "product": product,
                "quantity": quantity
            })
            product["stock"] -= quantity
            print(f'{product["name"]} x{quantity} added to cart.')
            return

    print("Invalid product ID!")

def remove_from_cart():
    if not cart:
        print("Cart is empty!")
        return

    print("\nItems in Cart:")
    for i, item in enumerate(cart):
        print(f"{i+1}. {item['product']['name']} x{item['quantity']}")

    try:
        index = int(input("Enter item number to remove: ")) - 1
    except ValueError:
        print("Invalid input!")
        return

    if 0 <= index < len(cart):
        removed = cart.pop(index)
        removed["product"]["stock"] += removed["quantity"]
        print("Item removed from cart.")
    else:
        print("Invalid choice!")

def checkout():
    if not cart:
        print("Cart is empty!")
        return

    print("\nEnter Customer Details")
    name = input("Enter Name: ")
    mobile = input("Enter Mobile Number: ")


    subtotal = 0
    total_discount = 0

    print("\n========== BILL ==========")
    print("Customer Name :", name)
    print("Mobile :", mobile)
    print("-" * 50

    for item in cart:
        product = item["product"]
        quantity = item["quantity"]

        discount = calculate_discount(product)
        price_after_discount = product["mrp"] - discount
        total_price = price_after_discount * quantity

        print(f'{product["name"]} x{quantity} | ₹{total_price:.2f}')

        subtotal += total_price
        total_discount += discount * quantity
    grand_total = subtotal - total_discount

    print("-" * 50)
    coupon = input("Enter coupon code (or press Enter): ")
    if coupon == VALID_COUPON:
        coupon_discount = grand_total * 0.10
        grand_total -= coupon_discount
        print("Coupon Applied! Extra 10% discount.")
    else:
        coupon_discount = 0

    print("-" * 50)
    print(f"Product Discount: ₹{total_discount:.2f}")
    print(f"Coupon Discount: ₹{coupon_discount:.2f}")
    print(f"Total Payable: ₹{grand_total:.2f}")
    print("=" * 50)
    print("Thank You for Shopping!")

    cart.clear()

while True:
    print("\n1. Show Products")
    print("2. Search Product")
    print("3. Add to Cart")
    print("4. Remove from Cart")
    print("5. Checkout")
    print("6. Exit")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input!")
        continue

    if choice == 1:
        show_products()
    elif choice == 2:
        search_product()
    elif choice == 3:
        add_to_cart()
    elif choice == 4:
        remove_from_cart()
    elif choice == 5:
        checkout()
    elif choice == 6:
        print("Exiting...")
        break
    else:
        print("Invalid choice!")
```
