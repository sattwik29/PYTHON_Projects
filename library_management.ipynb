```python
book_ids = []
book_names = []
authors = []
quantities = []
issued_books = []

def add_book():
    book_id = input("Enter Book ID: ")

    if book_id in book_ids:
        print("❌ Book ID already exists!")
        return

    name = input("Enter Book Name: ")
    author = input("Enter Author Name: ")
    qty = int(input("Enter Quantity: "))

    if qty <= 0:
        print("❌ Quantity must be greater than 0!")
        return

    book_ids.append(book_id)
    book_names.append(name)
    authors.append(author)
    quantities.append(qty)
    issued_books.append(0)

    print("✅ Book Added Successfully!\n")

def display_books():
    if len(book_ids) == 0:
        print("No books available!")
        return

    print("\n" + "="*90)
    print("ID\tName\tAuthor\tAvailable\tIssued")
    print("="*90)

    for i in range(len(book_ids)):
        available = quantities[i] - issued_books[i]
        print(f"{book_ids[i]}\t{book_names[i]}\t{authors[i]}\t{available}\t\t{issued_books[i]}")

    print("="*90)

def search_book():
    book_id = input("Enter Book ID to Search: ")

    if book_id in book_ids:
        i = book_ids.index(book_id)
        available = quantities[i] - issued_books[i]

        print("\n📖 Book Found!")
        print("Book Name:", book_names[i])
        print("Author:", authors[i])
        print("Total Quantity:", quantities[i])
        print("Available Copies:", available)
        print("Issued Copies:", issued_books[i])
    else:
        print("❌ Book Not Found!")

def issue_book():
    book_id = input("Enter Book ID to Issue: ")

    if book_id in book_ids:
        i = book_ids.index(book_id)
        available = quantities[i] - issued_books[i]

        if available > 0:
            issued_books[i] += 1
            print("✅ Book Issued Successfully!")
        else:
            print("❌ No copies available!")
    else:
        print("❌ Book Not Found!")

def return_book():
    book_id = input("Enter Book ID to Return: ")

    if book_id in book_ids:
        i = book_ids.index(book_id)

        if issued_books[i] > 0:
            issued_books[i] -= 1
            print("✅ Book Returned Successfully!")
        else:
            print("❌ No issued copies to return!")
    else:
        print("❌ Book Not Found!")

def delete_book():
    book_id = input("Enter Book ID to Delete: ")

    if book_id in book_ids:
        i = book_ids.index(book_id)

        del book_ids[i]
        del book_names[i]
        del authors[i]
        del quantities[i]
        del issued_books[i]

        print("✅ Book Deleted Successfully!")
    else:
        print("❌ Book Not Found!")


while True:
    print("\n📚 LIBRARY MANAGEMENT MENU")
    print("1. Add Book")
    print("2. Display Books")
    print("3. Search Book")
    print("4. Issue Book")
    print("5. Return Book")
    print("6. Delete Book")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        display_books()
    elif choice == "3":
        search_book()
    elif choice == "4":
        issue_book()
    elif choice == "5":
        return_book()
    elif choice == "6":
        delete_book()
    elif choice == "7":
        print("Exiting Library System...")
        break
    else:
        print("Invalid Choice!")
```
